Ques-3 Find the most frequent value and their indice(s) in the following arrays:

i. x = np.array([1,2,3,4,5,1,2,1,1,1])

```python
x=np.array([1,2,3,4,5,1,2,1,1,1])
ct_x=np.bincount(x)
print("frequency of each number:",ct_x)
mostfreq=np.argmax(ct_x)
print("most frequent number is:",mostfreq)
indices=np.where(x==mostfreq)[0]
print("indices of most freq number:",indices)
```

```python
y=np.array([1,1,1,2,3,4,2,4,3,3])
ct_y=np.bincount(y)
print("frequency of each number:",ct_y)
mostfreq_y=np.argmax(ct_y)
print("most frequent number is:",mostfreq_y)
indices_y=np.where(y==mostfreq_y)[0]
print("indices of most freq number:",indices_y)
```

Ques-4 For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'), find

i. Sum of all elements

ii. Sum of all element’s row-wise

iii. Sum of all element’s column-wise



```python
gfg=np.matrix('[4,1,9; 12,3,1; 4,5,6]')
print(gfg)
```

```python
sum_all_ele=np.sum(gfg)
print("total sum:",sum_all_ele)
```

```python
row_wise=np.sum(gfg,axis=1)
print("row wise sum:",row_wise)
```

```python
column_wise=np.sum(gfg,axis=0)
print("column wise sum:",column_wise)
```
